```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error
from sklearn.utils import shuffle
from datetime import datetime
```

```python
X_train = pd.read_csv("./data/X_train_text.csv")
X_submission = pd.read_csv("./data/X_test_text.csv")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
        X_train.drop(['Score'], axis=1),
        X_train['Score'],
        test_size=0.2,
        random_state=0
    )
```

```python
X_train_processed = X_train.drop(columns=['Id'])
X_test_processed = X_test.drop(columns=['Id'])
X_submission_processed = X_submission.drop(columns=['Id'])
```

```python
tree_num = [_ for _ in range(5, 101, 5)]
model_score = []
model_mae = []

for i in tree_num:
    print("tree num:", i)
    start = datetime.now()
    model = RandomForestClassifier(n_estimators=i, class_weight='balanced_subsample', random_state=0, n_jobs = 4)
    model.fit(X_train_processed, Y_train)
    Y_test_predictions = model.predict(X_test_processed)
    score = accuracy_score(Y_test, Y_test_predictions)
    model_score.append(score)
    mae = mean_squared_error(Y_test, Y_test_predictions)
    model_mae.append(mae)
    end = datetime.now()
    print("score:", score)
    print("MAE = ", mae)
    print("running time:", end-start)
score_max = max(model_score)
mae_min = min(model_mae)
score_max_tree = tree_num[model_score.index(score_max)]
mae_min_tree = tree_num[model_mae.index(mae_min)]
print('max score: {}'.format(score_max), 'tree num: {}'.format(score_max_tree))
print('min MAE: {}'.format(mae_min), 'tree num: {}'.format(mae_min_tree))

plt.plot(tree_num, model_score,'r-', label = 'Score')
plt.show()
plt.plot(tree_num, model_mae, 'g-', label = 'MAE')
plt.show()
```

```python
depth = [_ for _ in range(30, 101, 5)]
model_score = []
model_mae = []

for i in depth:
    print("depth:", i)
    start = datetime.now()
    model = RandomForestClassifier(n_estimators=39, max_depth=i, class_weight='balanced_subsample', random_state=0, n_jobs = 4)
    model.fit(X_train_processed, Y_train)
    Y_test_predictions = model.predict(X_test_processed)
    score = accuracy_score(Y_test, Y_test_predictions)
    model_score.append(score)
    mae = mean_squared_error(Y_test, Y_test_predictions)
    model_mae.append(mae)
    end = datetime.now()
    print("score:", score)
    print("MAE = ", mae)
    print("running time:", end-start)
score_max = max(model_score)
mae_min = min(model_mae)
score_max_tree = depth[model_score.index(score_max)]
mae_min_tree = depth[model_mae.index(mae_min)]
print('max score: {}'.format(score_max), 'depth: {}'.format(score_max_tree))
print('min MAE: {}'.format(mae_min), 'depth: {}'.format(mae_min_tree))

plt.plot(depth, model_score, 'r-', label = 'Score')
plt.show()
plt.plot(depth, model_mae, 'g-', label = 'MAE')
plt.show()
```

```python
model = RandomForestClassifier(n_estimators=30, max_depth=95, class_weight='balanced_subsample', random_state=0, n_jobs=4)
model.fit(X_train_processed, Y_train)

Y_test_predictions = model.predict(X_test_processed)

print("Accuracy on testing set = ", accuracy_score(Y_test, Y_test_predictions))
print("MAE = ", mean_squared_error(Y_test, Y_test_predictions))

cm = confusion_matrix(Y_test, Y_test_predictions, normalize='true')
sns.heatmap(cm, annot=True)
plt.title('Confusion matrix of the classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
model.fit(pd.concat([X_train_processed, X_test_processed], ignore_index=True), pd.concat([Y_train, Y_test], ignore_index=True))
X_submission['Score'] = model.predict(X_submission_processed)
```
